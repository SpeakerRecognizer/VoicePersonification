```python
import torch
import torchaudio

from VoicePersonification.models.itmo_personification_model_segmentation import (
    ITMOPersonificationModelSegmentation
)
```

```python
segmentator = ITMOPersonificationModelSegmentation()
```

Для демонстации работы сегментатора мы сконкатенируем две записи с разными дикторами.

```python
wav1, sr = torchaudio.load("./examples/id10289-3g9CjhcNEWk-00009.wav")
wav2, sr = torchaudio.load("./examples/id10277-znxUWA2QAGs-00001.wav")

wav = torch.concat([wav1, wav2], dim=-1)

assert sr == 16_000, "wav sample rate should be 16KGz"

embeddings, time_segments = segmentator.model.segmentate_chunk(wav.view(-1))

display(time_segments)
display(torch.cosine_similarity(embeddings[:-1], embeddings[1:]))
```

На выходе модели сегментации мы получили список сегментов, который состоит из пар (начало, конеч) в секундах и список эмбеддингов по одному для кажджого сегмента времени.
